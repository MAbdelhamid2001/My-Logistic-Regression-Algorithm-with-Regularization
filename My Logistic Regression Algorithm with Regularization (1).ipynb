```python
import numpy as np # for numerical operations
import pandas as pd # for handling input data
import matplotlib.pyplot as plt # for data visualization 
import seaborn as sns # for data visualization
data = pd.read_csv('ltrain.csv') # uses pandas library to open a .csv file
test_data = pd.read_csv('ltest.csv')
sample = pd.read_csv('lsample_submission.csv')
```

```python
#data=data.sample(frac=1)
size=int(0.7*len(data))
train=data[:size]
test=data[size:]
```

```python
xtrain=train.drop('genre_label',axis=1)
ytrain=train['genre_label']
```

```python
xtrain
```

```python
def scale(x):
    return ((x-np.mean(x))/np.std(x))
```

```python
x=scale(xtrain)
x=np.array(x)
y=np.array(ytrain)
w=np.ones((1,xtrain.shape[1]))
print("x shape",x.shape,"y shape",y.shape,"w shape",w.shape)
```

```python
def hypo(w,x,b):#prediction
    z=np.dot(w,x.T)+b#
    p_hat=1/(1+np.exp(-z))#g(z)    
    return p_hat
```

```python
def predict(p_hat,m):
    y_pred=np.zeros((1,m))
    for i in range(m):
        if p_hat[0][i]>=0.5:
            y_pred[0][i]=1
    return y_pred # represents h(x)
```

```python
def error(pred,y,m,lamda,w):#cost
    #costs= -np.sum(y * np.log(pred ) + (1 - y) * np.log(1 - pred )) / m
    epsilon = 1e-5 # error happend cause log doesnt accept 0 so i add 1e-5 to 0
    cost = -(1/(2*m))*(np.sum(y * np.log(pred + epsilon) + (1 - y) * np.log(1 - pred + epsilon))+lamda*np.sum(np.square(w)))
    return cost
```

```python
def gradient(pred,y,m,alpha,w,b,lamda,x):#gradient decent
    db=(1/m)*np.sum(pred-y.T)
    b=b-alpha*db
    w=w*(1-(alpha*lamda)/m)-(alpha/m)*np.dot((pred-y.T),x)#accoding to the equation of regularization
    return b,w
```

```python
#main
m=len(y)
alpha=0.05
lamda=0.01
b=1
cost=[]
for i in range(200000):
    p_hat= hypo(w,x,b)
    pred=predict(p_hat,m)
    er=error(pred,y,m,lamda,w)
    b,w= gradient(pred,y,m,alpha,w,b,lamda,x)
    if i%10000==0:
        print(er)
        cost.append(er)
```

```python
plt.plot(cost)
plt.ylabel('cost')
plt.xlabel('iterations')
plt.title('Cost reduction over time')
plt.show()
```
